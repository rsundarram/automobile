#### Sorting & Filter

```python
# sorting in ascending order
auto_price['engine-size'].sort_values()
```

```python
# All rows will be shuffled by engine-size in ascending order
auto_price.sort_values(by = 'engine-size', ascending = True)
```

```python
# descending order
auto_price.sort_values(by = 'engine-size', ascending = False)
```

```python
# unique values in each column
auto_price['make'].unique()
```

```python
# multiple columns 
auto_price[['make','body-style','engine-size']]
```

```python
auto_price
```

#### Filter DataFrame

```python
auto_price['make'] == 'audi'
```

```python
# Boolean Conditions
auto_price[auto_price['make'] == 'audi']
```

```python
# Multiple Conditions
auto_price[(auto_price['make'] == 'toyota') & (auto_price['num-of-doors'] == 'two')]
```

```python
auto_price[(auto_price['make'] == 'bmw') | (auto_price['num-of-doors'] == 'two')]
```

#### loc & iloc

```python
# slicing rows & columns 
# All rows & all columns
auto_price.loc[:,:]
```

```python
auto_price.loc[10:20,:]
```

```python
# iloc = slicing by using index position
# start index: stop index - 1
auto_price.iloc[10:20,:]
```

```python
# loc - columns names to select columns sequentially
auto_price.loc[10:20,'make':'engine-size']
```

```python
# start position : end position-1 : steps
auto_price.iloc[10:20,2:9]
```

```python
# Reverse all the column position 
# start index:stop index:step index
auto_price.iloc[10:20,::-1]
```

#### Sampling

```python
# n - number of samples
df_rand = auto_price.sample(n = 10)
df_rand
```

```python
# Frac = random sampling using fraction (%)
df_rand = auto_price.sample(frac = 0.1)
df_rand
```

#### Filter & Query

```python
# using filter method to filter columns 
# select columns by alphabet 'a' present in column names
auto_price.filter(like = 'a')
```

```python
# regrex = regular expression 
auto_price.filter(regex = '[an]')
```

```python
auto_price.query('make == "toyota" and `engine-location` == "front" and `engine-size` > 100') 
```

#### Drop
    * To Drop Rows / Columns

```python
# inplace = True , To update the drop columns in auto_price column
auto_price.drop(['symboling','normalized-losses'], axis = 1, inplace = True)
```

```python
auto_price.head()
```

```python
# Drop Rows , default : axis = 0
auto_price.drop([2, 3, 4, 5, 6, 7, 8], axis = 0)
```

#### Duplicate Rows

```python
# Returns True if duplicate rows are found
auto_price.duplicated()
```

```python
# No duplicate rows 
auto_price[auto_price.duplicated()]
```

```python
# Dropping duplicate rows
auto_price.drop_duplicates(inplace = True)
```

#### Missing Values

#### Groupby

```python
auto_price.groupby(['make'])
```

```python
auto_price.groupby(['make']).first()
```

```python
auto_price.groupby(['make','body-style']).first()
```

#### Concat

```python
data_series1 = pd.Series([10, 20, 30, 40, 50, 60, 70])
              
data_series2 = pd.Series([100, 200, 300, 400, 500, 600, 700])
```

```python
# Concatinate two Rows
pd.concat([data_series1,data_series2])
```

```python
# Concatinate two columns
pd.concat([data_series1,data_series2], axis = 1)
```
